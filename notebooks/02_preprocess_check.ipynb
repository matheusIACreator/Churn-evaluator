```python
import sys
sys.path.append('../src')

from preprocess import load_and_clean, encode, split, save_columns

df_raw =load_and_clean('../data/telco_churn.csv')
print('Shape bruto: ',df_raw.shape)
print('Nulos após limpeza: ', df_raw.isnull().sum().sum())
```

```python
df_encoded=encode(df_raw)
print("Shape após encode:", df_encoded.shape)
print("Tipos restantes:\n", df_encoded.dtypes.value_counts())
print("\nAinda tem object?", (df_encoded.dtypes == 'object').any())
```
